```python
data['Checkout_Duration'] = pd.to_numeric(data['Checkout_Duration'], errors='coerce')
data['Checkout_Click'] = pd.to_numeric(data['Checkout_Click'], errors='coerce')

task_1_duration_mean = data['Task_1_Duration'].mean()
task_2_item_1_duration_mean = data['Task_2_Item_1_Duration'].mean()
task_2_item_2_duration_mean = data['Task_2_Item_2_Duration'].mean()

task_1_click_mean = data['Task_1_Click'].mean()
task_2_item_1_click_mean = data['Task_2_Item_1_Click'].mean()
task_2_item_2_click_mean = data['Task_2_Item_2_Click'].mean()

task_1_hover_mean = data['Task_1_Hover'].mean()
task_2_item_1_hover_mean = data['Task_2_Item_1_Hover'].mean()
task_2_item_2_hover_mean = data['Task_2_Item_2_Hover'].mean()

checkout_duration_mean = data['Checkout_Duration'].mean()
checkout_click_mean = data['Checkout_Click'].mean()

plt.figure(figsize=(12, 6))

tasks = ['Task 1 (Search)', 'Task 2 Item 1 (Explore)', 'Task 2 Item 2 (Explore)', 'Checkout']
mean_durations = [task_1_duration_mean, task_2_item_1_duration_mean, task_2_item_2_duration_mean, checkout_duration_mean]
plt.bar(tasks, mean_durations, color=['skyblue', 'lightgreen', 'lightcoral', 'yellowgreen'])

plt.title('Mean Task Durations Comparison')
plt.ylabel('Duration (seconds)')
plt.xlabel('Tasks')
plt.show()

plt.figure(figsize=(12, 6))

mean_clicks = [task_1_click_mean, task_2_item_1_click_mean, task_2_item_2_click_mean, checkout_click_mean]
plt.bar(tasks, mean_clicks, color=['skyblue', 'lightgreen', 'lightcoral', 'yellowgreen'])

plt.title('Mean Task Clicks Comparison')
plt.ylabel('Number of Clicks')
plt.xlabel('Tasks')
plt.show()

plt.figure(figsize=(12, 6))

mean_hovers = [task_1_hover_mean, task_2_item_1_hover_mean, task_2_item_2_hover_mean]
hover_tasks = ['Task 1 (Search)', 'Task 2 Item 1 (Explore)', 'Task 2 Item 2 (Explore)']
plt.bar(hover_tasks, mean_hovers, color=['skyblue', 'lightgreen', 'lightcoral'])

plt.title('Mean Task Hovers Comparison')
plt.ylabel('Number of Hovers')
plt.xlabel('Tasks')
plt.show()
```

```python
task_2_item_1_threshold = 30
task_2_item_1_percentage = (data['Task_2_Item_1_Duration'] > task_2_item_1_threshold).sum() / data.shape[0] * 100
task_2_item_1_under_30_percentage = 100 - task_2_item_1_percentage
print(f"task_2_item_1_under_30s_percentage {task_2_item_1_under_30_percentage} %")
task_2_item_2_threshold = 50
task_2_item_2_percentage = (data['Task_2_Item_2_Duration'] > task_2_item_2_threshold).sum() / data.shape[0] * 100
task_2_item_2_under_50_percentage = 100 - task_2_item_2_percentage
print(f"task_2_item_2_under_50s_percentage {task_2_item_2_under_50_percentage} %")

plt.figure(figsize=(8, 5))
plt.bar(['Task 2 Item 1', 'Task 2 Item 2'], [task_2_item_1_percentage, 0], color='lightgreen', label='Task 2 Item 1 Duration> 30s')
plt.bar(['Task 2 Item 1', 'Task 2 Item 2'], [task_2_item_1_under_30_percentage, 0], bottom=[task_2_item_1_percentage, 0], color='darkgreen', label='Task 2 Item 1 Duration <= 30s')
plt.bar(['Task 2 Item 1', 'Task 2 Item 2'], [0, task_2_item_2_percentage], color='lightcoral', label='Task 2 Item 2 Duration> 50s')
plt.bar(['Task 2 Item 1', 'Task 2 Item 2'], [0, task_2_item_2_under_50_percentage], bottom=[0, task_2_item_2_percentage], color='darkred', label='Task 2 Item 2 Duration <= 50s')

plt.title('Percentage of Participants by Task Duration')
plt.ylabel('Percentage of Participants (%)')
plt.ylim(0, 100)
plt.legend()
plt.show()
```

```python
data[['Task_1_Duration', 'Task_2_Item_1_Duration', 'Task_2_Item_2_Duration']].plot(kind='box')
plt.title('Task Duration Distribution')
plt.show()
```

```python
features = data[['Task_1_Duration', 'Task_2_Item_1_Duration', 'Task_2_Item_2_Duration',
                 'Task_1_Click', 'Task_2_Item_1_Click', 'Task_2_Item_2_Click',
                 'Task_1_Hover', 'Task_2_Item_1_Hover', 'Task_2_Item_2_Hover']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=2, random_state=36)  
kmeans.fit(scaled_features)

data['Cluster'] = kmeans.labels_


cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=features.columns)
cluster_centers_df
```
